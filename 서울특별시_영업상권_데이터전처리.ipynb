```python
def get_address(address):
    
    '''
    geopy 모듈을 이용하여 도로명 주소에서 경도, 위도를 담은
    자료로 변경해주는 함수 
    '''
    
    

    geolocoder = Nominatim(user_agent = 'South Korea', timeout = None)
    geo = geolocoder.geocode(address)
    
    return geo
```

# 53만개의 데이터를 모두 적용하기 보다, 각 유니크한 값의 경도, 위도를 추출 한 후 
# map 을 이용하여 변경해주도록 함 

```python
location = pd.Series(df['소재지전체주소'].unique())
address = location.progress_map(get_address)
address
```

```python
none_type_num = address[address.isna()].index #주소가 nonetype 인 경우의 인덱스 번호 저장 
none_type_value = location[none_type_num] #주소가 nonetype 인 경우의 주소 저장 
new_location = location.drop(index = none_type_num, axis = 0)

new_location
```

```python
none_type_index = df.loc[df['소재지전체주소'].isin(none_type_value)].index #주소가 nonetype인 데이터 프레임의 인덱스
df = df.drop(index = none_type_index, axis = 0) #주소가 nonetype인 데이터 프레임의 인덱스 제거 
df
```

```python
new_address = address.drop(index = none_type_num)
new_address
```

```python
lat = new_address.progress_map(get_lat)
lot = new_address.progress_map(get_lot)
```

```python
def make_dict(x,y):
    
    X = np.array(x).reshape(-1,1)
    Y = np.array(y).reshape(-1,1)
    
    return dict(np.c_[X,Y])
```

```python
dicted_lat = make_dict(new_location,lat)
dicted_lot = make_dict(new_location,lot)
dicted_lat
```

```python
df['lat'] = df['소재지전체주소'].map(dicted_lat)
df['lot'] = df['소재지전체주소'].map(dicted_lot)
df
```

### 확인 결과 일반음식점 제조업을 제외한 다른 것들은, 업태구분명이 모두 없는 것으로 판명되었다.
### 일반음식점은 데이터를 둘러보고 결측치를 변경해주고, 그 이상의 것들은 데이터의 특성에 따라 변경해주자 

```python
cols = df.loc[df['업태구분명'].isna(),'개방서비스명'].value_counts().index.tolist()[:-1]

for _ in cols:
    cond = (df['업태구분명'].isna()) & (df['개방서비스명'] == _)
    df.loc[cond,'업태구분명'] = df.loc[cond,'업태구분명'].fillna(df.loc[cond,'개방서비스명']) #개방서비스명으로 업태구분명을 채워줌
```

```python
df.loc[(df['업태구분명'].isna())& (df['개방서비스명'] == '일반음식점')] #각자 해당하는 업태구분명이 달라보인다.
```

```python
df.loc[df['개방서비스명'] == '일반음식점','업태구분명'].unique()
```
