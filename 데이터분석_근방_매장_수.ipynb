## 근방 매장 수 세기
* 500m 기준을 위도 0.0022로 잡아서 수집

```python
def count_num(a,b):
    row_list=[]
    for i, row in df.iterrows():
        if (a-0.0022 <= row['lat'] <= a+0.0022) and (b-0.0022 <= row['lot'] <= b+0.0022):
            row_list.append(row['매장'])
    return len(row_list)
```

```python
df['근방 매장수'] = pd.DataFrame({'A': df['lat'], 'B': df['lot']}).apply(lambda x : count_num(x['A'], x['B']), axis=1)
```

```python
df['근방 매장수'].describe()
```

```python
def count_num2(a,b):
    row_list=[]
    for i, row in df.iterrows():
        if (a-0.0044 <= row['lat'] <= a+0.0044) and (b-0.0044 <= row['lot'] <= b+0.0044):
            row_list.append(row['매장'])
    return len(row_list)
```

```python
df_n['근방 매장수_1km'] = pd.DataFrame({'A': df['lat'], 'B': df['lot']}).apply(lambda x : count_num2(x['A'], x['B']), axis=1)
```

## Minmax 정규화

```python
df['근방 매장수 정규화'] = df['근방 매장수'].apply(minmax)
```

## 근방 매장 수 기준

```python
df2 = df.groupby('구')['근방 매장수'].sum().reset_index()
df2 = df2.sort_values(by='근방 매장수', ascending=False)
x = df2['구']
y = df2['근방 매장수']
fig, ax = plt.subplots(figsize=(20, 5))
sns.barplot(x, y, ax=ax)
plt.show()
```

```python
df3 = df.groupby('구')['매장'].count().reset_index()
df3 = df3.sort_values(by='매장', ascending=False)
x = df3['구']
y = df3['매장']
fig, ax = plt.subplots(figsize=(20, 5))
sns.barplot(x, y, ax=ax)
plt.show()
```

```python
df4 = df.groupby('구')['근방 매장수 정규화'].sum().reset_index()
df4 = df4.sort_values(by='근방 매장수 정규화', ascending=False)
x = df4['구']
y = df4['근방 매장수 정규화']
fig, ax = plt.subplots(figsize=(20, 5))
sns.barplot(x, y, ax=ax)
plt.show()
```
