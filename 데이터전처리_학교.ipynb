## 중고등학교 - 위경도 추가

```python
import pandas as pd
df_s = pd.read_csv('/Users/hollyyoon/Downloads/서울시 학교 기본정보.csv', encoding='cp949')
df_s.head()
```

```python
df_s = df_s[['학교종류명','학교명','도로명우편번호', '도로명주소','도로명상세주소']]
df_s
```

```python
def find_lat(name):
    try :
        tmp = gmaps.geocode(name,language='ko')
        return tmp[0].get('geometry')['viewport']['northeast']['lat']
    except : 
        return None
```

```python
def find_lot(name):
    try : 
        tmp = gmaps.geocode(name,language='ko')
        return tmp[0].get('geometry')['viewport']['northeast']['lng']
    except : 
        return None
```

```python
df_s['lat'] = df_s['학교명'].apply(find_lat)
df_s['lot'] = df_s['학교명'].apply(find_lot)
df_s.head()
```

```python
df_s.to_csv('/Users/hollyyoon/Downloads/중고교.csv')
```
